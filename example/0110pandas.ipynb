```python
# Pandas:Series,DataFrame(Table)
# 樞紐分析表:欄位、列、中間區:值
# pivot
import pandas as pd

data = {
    '分類':['居家', '居家', '娛樂', '娛樂', '科技', '科技'],
    '商店':['家樂福', '頂好', '家樂福', '全聯', '頂好', '家樂福'],
    '價格':[11.42, 23.5, 19.99, 15.95, 55.75, 111.55],
    '測試分數':[4, 3, 5, 7, 5, 8]
}
data

products = pd.DataFrame(data)
products
```

```python
pivot_products = products.pivot_table(index = '分類', columns = '商店', values = '價格')
pivot_products
```

```python
pivot_products = products.pivot_table(index = '分類', columns = '商店', values = '測試分數')
pivot_products
```

```python
# linspace
import numpy as np
# 0-10之間產生50個
x = np.linspace(0, 10 , 50)
sinus = np.sin(x)
cosinus = np.cos(x)
plt.plot(x, sinus,'r-o', label = 'sin(x)' )
plt.plot(x, cosinus, 'g--', label = 'cosin(x)')
# 格線
plt.grid(True)

# 文字說明
plt.xlabel('red line value of sin\ngreen line value of cosin')
# 標題
plt.title('My science plot example')
# 圖說(在plot要先設定label)
# loc = 'center' or 'upper right'
plt.legend(loc = 'upper right')
plt.savefig('test.png')
plt.show()
```

```python
# 色彩: b, g, r , w, k
# 線條: -, -- , . ,
# 散點圖、散佈圖、scatter
x = np.linspace(0, 2*np.pi, 50)
y = np.sin(x)
plt.scatter(x,y)
plt.show()
```

```python
# 長條圖 bar -> 不同群，不同種類的資料大小
labels = ['Python', 'C++', 'Java', 'JS', 'C', 'C#']
# 刻度
index = np.arange(len(labels))
# 使用頻率

ratings = [5.16, 5.726, 14.988, 3.165, 11.854, 4.453]
# plt.bar(index, ratings)
# plt.xticks(index,labels)
# 重直長條圖
# plt.bar(labels, ratings)
# 水平長條圖
plt.barh(labels, ratings)
plt.ylabel('Use %')
plt.title('Most popular program language')
plt.show()
```

```python
# 長條圖 bar -> 不同群，不同種類的資料大小
labels = ['Python', 'C++', 'Java', 'JS', 'C', 'C#']
# 刻度
index = np.arange(len(labels)*2)
# 使用頻率

ratings = [5.16, 5.726, 14.988, 3.165, 11.854, 4.453]
change = [1.12, 0.3, -1.69, 0.29, 3.41, -0.45]
# index範圍值0-2
plt.bar(index[0::2], ratings, label='Use rate')
plt.bar(index[1::2], change, label = 'Groth rate', color = 'r')
plt.legend()
plt.xticks(index[0::2],labels)
# 重直長條圖
# plt.bar(labels, ratings)

# 水平長條圖
# plt.barh(labels, ratings)
plt.ylabel('Use %')
plt.title('Most popular program language')
plt.show()
```

```python
# 直方圖 histograms,顯示數值資料分布，次數分配
# 常態分配normal distribution
x = np.random.randn(1000)
# 50條直方圖
num_bins = 50
plt.hist(x, num_bins)
plt.show()
```

```python
# 餅圖，派圖(pie chart)顯示資料分布的%
labels = ['Python', 'C++', 'Java', 'JS', 'C', 'C#']
ratings = [5, 6, 15, 3, 12, 4]
# 使資料突出
explode = [0, 0, 0, 0, 0, 0.2]
plt.pie(ratings, labels = labels, explode = explode)
plt.title('Program Language Rating')
plt.show()
# boxplot,radiation....
```

```python
# 多軸折線
dists = {
    'Area':['中正區', '板橋區', '桃園區', '北屯區', '安南區', '三民區',
          '大安區', '永和區', '八德區', '前鎮區', '鳳山區', '信義區', '新店區'],
    'People':[159598, 551452, 441287, 275207, 192327, 343203, 309835, 222531, 198473, 189623, 359125, 225561, 302070],
    'Place':[7, 23, 34, 62, 107, 19, 11, 5, 33, 19, 26, 11, 120]
}

df = pd.DataFrame(dists, columns= ['People', 'Place'], index = dists['Area'])

print(df)

# 因為人口跟面積值相差很大直接繪圖會很奇怪，所以給面積成1000倍
df['Place'] *= 1000
# rot為兩相對值的角度
df.plot(xticks = range(len(df.index)),use_index=True, rot = 90)
plt.show()
```
