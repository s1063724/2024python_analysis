```python
#http://lib.stat.cmu.edu/datasets/boston

import numpy as np
import pandas as pd
# sep是區隔符號，s+ : 多個space
# skiprows:跳過前面22行
raw_df = pd.read_csv('boston.csv', sep='\s+', skiprows=22, header=None)
raw_df.head(10)
```

```python
# hstack水平堆疊(觀察後發現此資料每14筆為1串資料，但被分成了兩行)
# ::2最前面到最後隔2個抓一個
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
# 前面抓的是:0、2、4、6、8 ....整列資料
# 後面抓的是:1、3、5、7....前兩行的資料
# 目的將兩列資料合併為一列((第14筆為房屋均價，作為目標所以不抓))
target = raw_df.values[1::2, 2]# 第14筆資料(房屋均價)
print(data)
print(data.shape)
```
