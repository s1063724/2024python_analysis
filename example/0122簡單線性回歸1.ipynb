```python
# 高中生身高體重
# 身高: 147.9, 163.5, 159.8, 155.1, 163.3, 158.7, 172.0, 161.2, 153.9, 161.6
# 體重: 41.7, 60.2, 47.0, 53.2, 48.3, 55.2, 58.5, 49.0, 46.7, 52.5
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
h = np.array([147.9, 163.5, 159.8, 155.1, 163.3, 158.7, 172.0, 161.2, 153.9, 161.6])
w = np.array([41.7, 60.2, 47.0, 53.2, 48.3, 55.2, 58.5, 49.0, 46.7, 52.5])
```

```python
X = pd.DataFrame(h, columns=['身高'])
y = pd.DataFrame(w, columns=['體重'])
lm = LinearRegression()
lm.fit(X,y)
print('迴歸係數',lm.coef_)
print('截距',lm.intercept_)
```

```python
# 預測身高 150, 160, 170的高中生體重
new_h = pd.DataFrame(np.array([150,160,177]),columns = ['身高'])
pred_w = lm.predict(new_h)
print('AI 的答案:\n',pred_w)
```

```python
plt.scatter(h, w)
re_w = lm.predict(X)
plt.plot(h, re_w, color = 'b')
plt.plot(new_h, pred_w, color = 'r', marker = 'o', markersize = 10)
plt.title('Proper Weight for High School Student')
plt.show()
```
